Defining and preprocessing input-output features:

let's define the input (X) and output (y) features separately.

```python
X = df.drop(labels= "price_range" , axis = 1)
y = df["price_range"]
# X.head()
```

After slicing, let’s now create a pair of training and testing sets by using the train_test_split() function which will be used in the training and testing phase respectively. We are keeping 70% of the data for training and the rest 30% will be used to evaluate the classifier.   

### Without Scaling

First let us look,how our model performs without using any scaling techniques

```python
#Splitting into test and train set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
NNH = KNeighborsClassifier(n_neighbors= 45 ,metric='euclidean' )
```

```python
NNH.fit(X_train, y_train)
```

```python
NNH.score(X_train, y_train)
```

```python
NNH.score(X_test, y_test)
```

From the above accuracies for training and testing, the classifier has learned the distances very well between the training data points, and also it has also generalized very well on testing data points.  

### With Standard Scaler

As there are few features that have values higher than other features and they can dominate the prediction result hence we need scaled-down values for all features in the same range. For this purpose, we are using the StandardScaler function from the sklearn library and will scale both training and testing patterns. 

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.preprocessing import StandardScaler
scalar1 = StandardScaler()
x1_train_scaler=scalar1.fit_transform(X_train)
x1_test_scaler=scalar1.fit_transform(X_test)
```

```python
NNH = KNeighborsClassifier(n_neighbors= 45 ,metric='euclidean' )
```

```python
NNH.fit(x1_train_scaler, y_train)
```

```python
NNH.score(x1_train_scaler, y_train)
```

```python
NNH.score(x1_test_scaler, y_test)
```

With standardscaler technique the model does not perform very well, this may be due to as the datapoints are scaled down to very small values,that becomes difficult for the model to identifies its neighbours
