##Building the model 

###1. Linear Regression

```python
#implementing Linear Regression
X = hp.drop(['Property_Sale_Price'], axis=1)
#Feature selection
# the dependent variable 
y= hp[['Property_Sale_Price']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python

# fit the model
reg_model = LinearRegression()
reg_model.fit(X_train, y_train)
```

```python
reg_model.score(X_train, y_train),reg_model.score(X_test, y_test)
```

Here, the training-set accuracy score is **0.8694** while the test-set accuracy to be **0.7419**.
The model is overfitting as in training data the model performs well on the training data but does not perform well on the evaluation data.
